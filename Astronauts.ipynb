# Astronauts Dataset

Taken from [International Astronaut Database](https://aerospace.csis.org/data/international-astronaut-database/).

(Latest dataset download 15/06/2020)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('International Astronaut Database.csv')
```

```python
df.head()
```

##### Replace "Soviet Union" with "Russia"

```python
df.replace('Soviet Union', 'Russia', inplace=True)
```

```python
df[df.Country == 'Soviet Union']
```

```python
f, ax = plt.subplots(figsize=(10, 15))
sns.countplot(y="Country", data=df, palette="Set3", order = df['Country'].value_counts().index)
```

```python
df_except = df[~df['Country'].isin(['United States', 'Russia'])]
```

```python
f, ax = plt.subplots(figsize=(10, 15))
sns.countplot(y="Country", data=df_except, palette="Set3", order = df_except['Country'].value_counts().index);
```

```python
sns.catplot(x="Gender", kind="count", data=df);
```

```python
df_w = df[df['Gender'] == 'Woman']
df_w
```

```python
f, ax = plt.subplots(figsize=(10, 15))
sns.countplot(y="Country", data=df_w, palette="Set3", order = df_w['Country'].value_counts().index)
```
